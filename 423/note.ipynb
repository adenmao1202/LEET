#### shortest path 
1. unweighted 
    - bfs
    - define all edges to be 1 
    - shortest path : use min edges 
    - time: O(n + m ) 
    - n : # of vertices
    - m : # of edges

2. weighted 
    - bfs : not possible
    - dijkstra

3. negative weights
    - shortest path problem beconme meaningless --> since we can infinite loop to decrease cost 
    - Hence, we assume no negative weights --> pos weights
