```python
def cut_str(cuts:list, m:int, n:int)->int:
    p = len(cuts)  # here total p-2 cut points
    c = [x[:] for x in [[0]*p]*p] # initial
    for r in range(2, p): # cut range from 2 to len(cuts)-1
        for i in range(p-r): # i is start boundary
            j = i + r   # fill c[i][j] order for p=5: [0][2],[1][3],[2][4]; [0][3],[1][4]; [0][4]
            # 根據遞迴關係式
            min_sub = min(c[_][_]+c[_][_] for k in range(i+1,j)) # need fix: 找出成本最小的切斷點
            c[i][j] = cuts[_]-cuts[_]+min_sub # need fix: 切的字串總長度+上面找的切斷處最小成本
    return c[m][n]
```

```python
cuts = [0,3,8,10,20]
print(cut_str(cuts, 0, len(cuts)-1))
#37
cuts = [0,3,8,10,14,20]
print(cut_str(cuts, 0, len(cuts)-1))
#46
cuts = [0,9,11,15,18,44,67,101]
print(cut_str(cuts, 0, len(cuts)-1))
#235
cuts = [0,20]
print(cut_str(cuts, 0, len(cuts)-1))
#0
cuts = [0,8,10,14,29,41,59,62,76,79,91,102,105,106,135,142,147,149,150,153,161,163,168,173,188,189,193,199,200,201,209,216,219,226,239,246,253,272,284,286,292,299,315,316,317,333]
print(cut_str(cuts, 0, len(cuts)-1))
#1724
```

```python
def is_valid_subset_sum(s:list, T:int)->bool:
    M = [x[:] for x in [[False]*(T+1)]*(len(s)+1)]
    for i in range(1,len(s)+1):
        for j in range(1,T+1):

            M[i][j]  = (j == s[i-1]) or ((M[i-1][j-s[i-1]] or M[i-1][j]) if j > s[i-1] else M[i-1][j])
            # M[i][j]內容= 遞迴關係2: j是否等於s[i-1](i比索引多1) 或
            # 遞迴關係3: 當j>s[i-1]時 上一列M[i-1][j-s[i-1]]或M[i-1][j]是否為真
            # 遞迴關係1: 不然直接複製上一列的內容M[i-1][j]

    return M[__][__] #need fix: 回傳哪一格的值？
```

```python
print(is_valid_subset_sum([1,9,2,10,5],22))
#True
print(is_valid_subset_sum([1,9,2,10,5],23))
#False
print(is_valid_subset_sum([],1))
#False
print(is_valid_subset_sum([14,39,50,25,51,20,22,40,17,13,56,15,5,60,30],399))
#True
print(is_valid_subset_sum([17,31,56,41,23,60,5,24,26,42,8,45,14,2,15,7],398))
#False
```
